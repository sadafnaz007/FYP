```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the dataset
file_path = 'dataset.csv'
data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# Display the first few rows of the dataset and the data types of each column
print(data.head())
print("\n")
print("****Datatypes are as follows****") 
print(data.dtypes)
```

```python
# Filter the data to show from the start till 30th November,2022
filtered_data = data[:'2022-11-30']

# Plotting the filtered data as a line chart
plt.figure(figsize=(10, 8))
plt.plot(filtered_data['Close'], label='Line Chart of Close Price', color='purple')
plt.title('Line Chart of Close Price Trend until November 2022')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filter the data to show from 1st December,2022 till 31st December,2022
next_filtered_data = data['2022-12-01':'2022-12-31']

# Plotting the filtered data as a line chart
plt.figure(figsize=(10, 8))
plt.plot(next_filtered_data['Close'], label='Line Chart of Close Price', color='red')
plt.title('Line Chart of Close Price Trend until November 2022')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
